# Housing Price Analysis

### Import Necessary Libraries

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from matplotlib.ticker import FuncFormatter
```

### Import Data

```python
df = pd.read_csv("train.csv")
df
```

### Summary Statistics

#### Numerical Features

```python
numerical_features = df.select_dtypes(include=[np.number])
num_summary_stats = numerical_features.describe().T
num_summary_stats
```

#### Categorical Features

```python
categorical_features = df.select_dtypes(include=[object])
cat_summary_stats = categorical_features.describe().T
cat_summary_stats
```

### Evaluate Null Values

#### Features with Missing Values

```python
null_values = df.isnull().sum()
missing_percentage = (null_values / len(df)) * 100

missing_data = {'Null Values Count': null_values, 'Missing Percentage': missing_percentage}
missing_data_df = pd.DataFrame(missing_data)

missing_data_df = missing_data_df.loc[missing_data_df['Null Values Count'] > 0]
missing_data_df
```

#### Rows with Missing Values

```python
rows_with_missing_values = df[df.isnull().any(axis=1)]
rows_with_missing_values
```

### Explore Data

```python
def currency_formatter(value, pos):
    if value >= 1e3:
        value /= 1e3
        return f'${value:.0f}K'
    return f'${value:.0f}'

usd_formatter = FuncFormatter(currency_formatter)
```

#### 1. What is the distribution of dwelling types and how do they relate to sale prices?

```python
dwelling_types_counts = df['BldgType'].value_counts()
dwelling_prices = df.groupby('BldgType')['SalePrice'].mean()

dwelling_data = {'Count': dwelling_types_counts, 'AveragePrice': dwelling_prices}
dwelling_data_df = pd.DataFrame(dwelling_data)
# dwelling_data_df.index.name = None
dwelling_data_df.reset_index(inplace=True)

dwelling_data_display = dwelling_data_df.copy()
dwelling_data_display['AveragePrice'] = dwelling_data_display['AveragePrice'].map('${:,.2f}'.format)

dwelling_data_display
```

```python
dwelling_data_df = dwelling_data_df.sort_values('Count', ascending=False)

sns.set_style("darkgrid")

ax = sns.barplot(x="BldgType", y="Count", data=dwelling_data_df, palette="Set2")
plt.xlabel('Building Type', labelpad=15)
plt.ylabel('Count', labelpad=15)
plt.title('Dwelling Types Count', size=15, pad=20)

"""
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center',
                fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')
"""

plt.tight_layout()
plt.show()
```

```python
# formatted_dwelling_prices = ['$' + f'{value:,.2f}' for value in dwelling_prices.values]
dwelling_data_df = dwelling_data_df.sort_values('AveragePrice', ascending=False)

ax = sns.barplot(x="BldgType", y="AveragePrice", data=dwelling_data_df, palette="Set3")
plt.xlabel('Building Type', labelpad=15)
plt.ylabel('Price', labelpad=15)
plt.title('Average Sale Price', size=15, pad=20)

ax.yaxis.set_major_formatter(usd_formatter)

plt.tight_layout()
plt.show()
```

#### 2. Is there a correlation between the age of the property and its sale price?

```python
df['PropertyAge'] = df['YrSold'] - df['YearBuilt']

# If age_price_corr is close to 1, it indicates a strong positive correlation (as property age increases, sale price tends to increase), 
# while if it's close to -1, it indicates a strong negative correlation (as property age increases, sale price tends to decrease). 
# A value close to 0 suggests little to no linear correlation.
age_price_corr = df['PropertyAge'].corr(df['SalePrice'])
print(f'Correlation between Property Age and Sale Price: {age_price_corr}')

property_age_sale_price_df = df[['PropertyAge', 'SalePrice']]
property_age_sale_price_df = property_age_sale_price_df.sort_values('PropertyAge')
property_age_sale_price_df.reset_index(drop=True, inplace=True)

property_age_sale_price_display = property_age_sale_price_df.copy()
property_age_sale_price_display['SalePrice'] = property_age_sale_price_display['SalePrice'].map('${:,.2f}'.format)

property_age_sale_price_display
```

```python
plt.style.use("dark_background")

ax = sns.scatterplot(x="PropertyAge", y="SalePrice", data=property_age_sale_price_df, hue="PropertyAge", palette="GnBu")

plt.xlabel('Property Age', labelpad=15)
plt.ylabel('Price', labelpad=15)
plt.title('Property Age vs Sale Price', size=15, pad=20)

ax.yaxis.set_major_formatter(usd_formatter)

legend = plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
legend.set_title("Age")

plt.tight_layout()
plt.show()
```

#### 3. Is there a correlation between the living area size and the sale price?

```python
living_area_price_corr = df['GrLivArea'].corr(df['SalePrice'])
print(f'Correlation between Living Area (above grade) and Sale Price: {living_area_price_corr}')

living_area_sale_price_df = df[['GrLivArea', 'SalePrice']]
living_area_sale_price_df = living_area_sale_price_df.sort_values('GrLivArea')
living_area_sale_price_df.reset_index(drop=True, inplace=True)

living_area_sale_price_display = living_area_sale_price_df.copy()
living_area_sale_price_display['SalePrice'] = living_area_sale_price_display['SalePrice'].map('${:,.2f}'.format)

living_area_sale_price_display
```

```python
ax = sns.scatterplot(x="GrLivArea", y="SalePrice", data=living_area_sale_price_df, hue="GrLivArea", palette="GnBu")

plt.xlabel('Living Area (above grade)', labelpad=15)
plt.ylabel('Price', labelpad=15)
plt.title('Living Area (above grade) vs Sale Price', size=15, pad=20)

ax.yaxis.set_major_formatter(usd_formatter)

legend = plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
legend.set_title("Area")

plt.tight_layout()
plt.show()
```
